```python
# Scenario:
# We want to classify whether a restaurant order is “Large” (1) or “Small” (0) based on the number
#  of items ordered.
# - If items ≥ 3 → Large order
# - If items < 3 → Small order

import numpy as np

x=np.array([[3],[5],[4],[1],[2],[6]])
y=np.array([1,1,1,0,0,1])

#weights and bias 
weight=np.random.rand(1)
bias=np.random.rand(1)
learning_rate=0.1

def activation(z):
    return 1 if z >= 3 else 0
    
for epoch in range(10):  # run for 10 iterations
    print(f"Epoch {epoch+1}")
    for i in range(len(x)):
        # Linear combination
        z = np.dot(x[i], weight) + bias
        y_pred = activation(z)

        # Error
        error = y[i] - y_pred

        # Update rule
        weight += learning_rate * error * x[i]
        bias += learning_rate * error

        print(f"Input: {x[i]}, Predicted: {y_pred}, Actual: {y[i]}, Error: {error}")

print("\nFinal Weights:", weight)
print("Final Bias:", bias)

# Step 5: Test the perceptron
test_hours = np.array([3, 7])
for h in test_hours:
    z = np.dot(h, weight) + bias
    print(f"no of items: {h}, Prediction: {activation(z)}")
    
    
```
