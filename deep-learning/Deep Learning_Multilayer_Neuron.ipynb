```python
import numpy as np

X = np.array([
    [2,15],
    [3,20],
    [8,60],
    [9,75]
])

y = np.array([[0],[0],[1],[1]])

X = X / np.max(X, axis=0)

np.random.seed(0)

W1 = np.random.randn(2,3)
b1 = np.zeros((1,3))

W2 = np.random.randn(3,1)
b2 = np.zeros((1,1))

def relu(x):
    return np.maximum(0,x)

def sigmoid(x):
    return 1/(1+np.exp(-x))

lr = 0.1
epochs = 5000

for i in range(epochs):

    Z1 = np.dot(X,W1) + b1
    A1 = relu(Z1)

    Z2 = np.dot(A1,W2) + b2
    y_pred = sigmoid(Z2)

    loss = np.mean((y - y_pred)**2)

    dZ2 = y_pred - y
    dW2 = np.dot(A1.T,dZ2)
    db2 = np.sum(dZ2,axis=0)

    dA1 = np.dot(dZ2,W2.T)
    dZ1 = dA1 * (Z1 > 0)

    dW1 = np.dot(X.T,dZ1)
    db1 = np.sum(dZ1,axis=0)

    W1 -= lr*dW1
    b1 -= lr*db1
    W2 -= lr*dW2
    b2 -= lr*db2

print("Predictions:")
print(y_pred)
```
